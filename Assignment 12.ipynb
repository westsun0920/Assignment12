```python
import numpy as np
import matplotlib.pyplot as plt

num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (clean) y-coordinate data 
# y2 : (noisy) y-coordinate data

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x) 			# clean points
y2      = y1 + nn * std		# noisy points

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

```python
def leastSquare(v1,v2):
    v1 = np.mat(v1)
    v2 = np.mat(v2).T
    v3 = v1.T * v1
    theta = v3.I * (v1.T * v2) # weight
    return theta
```

```python
def computeLambda(matrix, vec, lambda_):
    column = len(matrix.T)
    lamvec = np.ones((1,column), dtype=float)
    zerovec = [0]
    matrixA = np.concatenate((matrix, lambda_*lamvec), axis=0)
    matrixY = np.concatenate((vec, zerovec), axis=0)
    return matrixA, matrixY
```

```python
def computefunction(n , lam):
    dimensions = []; dimenX = []; var = []; weight = []; y = []; y_ = []
    for i in range(0, len(x)):
        dimenX = []
        for j in range(0, n+1):
            dimenX.append(x[i]**j)
        dimensions.append(dimenX)  # Set X0 = 1.0 for computing
    dimensions = np.mat(dimensions)
    matrixA, matrixY = computeLambda(dimensions, y2, lam)
    weight = leastSquare(matrixA, matrixY)
    for j in range(0, n+1):
        var = []
        var = (x**j)*(float(weight[j]))
        y_.append(var)
    y_ = np.mat(y_)
    y_ = y_.T
    for i in range(0, len(y_)):
        sum_ = np.sum(y_[i])
        y.append(float(sum_))
    return y
```

```python
def computeError(v1, v2):
    error = 0
    for i in range(0, len(v2)):
        error += np.sqrt((v1[i] - v2[i])**2)
    print("MSE: ", error)        
    return np.sqrt(error)
```
